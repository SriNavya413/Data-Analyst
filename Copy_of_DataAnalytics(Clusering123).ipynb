### **f.Consider Tufts University, which is missing some information. Compute the Euclidean distance of this record from each of the clusters that you found above (using only the measurements that you have). Which cluster is it closest to? Impute the missing values for Tufts by taking the average of the cluster on those measurements.**

```python
import pandas as pd, numpy as np
from sklearn.cluster import KMeans

# Load and clean data
df = pd.read_csv('/content/Universities.csv').replace('...', np.nan).dropna()
cols = df.columns[2:]
df[cols] = df[cols].astype(float)

# Normalize numeric data
X = (df[cols] - df[cols].mean()) / df[cols].std()

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=X.columns)

# Show all valid column names for reference
print("\nValid column names:\n", cols.tolist())

# Tufts record using matching column names only
tufts_data = {col: np.nan for col in cols}
tufts_data.update({
    'Public (1)/ Private (2)': 2,
    'in-state tuition': 59464,
    'out-of-state tuition': 59464,
    'board': 4886,
    'add. fees': 0,
    'estim. book costs': 1100,
    'estim. personal $': 2300,
    '% fac. w/ PHD': 0.961,
    'stud./fac. ratio': 9.9,
    'Graduation rate': 95.0
})

# Create Tufts DataFrame
tufts = pd.DataFrame([tufts_data])

# Normalize Tufts, compute distance to centroids
tufts_norm = (tufts - X.mean()) / X.std()
distances = [np.linalg.norm(tufts_norm - c) for _, c in centroids.iterrows()]
cluster_index = np.argmin(distances)

# Impute missing using nearest cluster centroid
tufts_imputed = tufts.fillna(centroids.iloc[cluster_index] * X.std() + X.mean()).round(0)
print("\nImputed Tufts Record:\n")
print(tufts_imputed)

```
