```
# prompt: Using dataframe data: 7 different graphs using seaborn

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' DataFrame is already loaded as in the previous code
# If not, load it here with the correct encoding
# data = pd.read_csv('/content/PoliceKillingsUS.csv', encoding='latin-1')

# Convert the 'date' column to datetime objects and extract the year
# This needs to be done within this cell if 'year' was created in a previous cell
data['date'] = pd.to_datetime(data['date'])
data['year'] = data['date'].dt.year

# 1. Countplot of manner of death
plt.figure(figsize=(8, 6))
sns.countplot(x='manner_of_death', data=data)
plt.title('Manner of Death')
plt.show()

# 2. Histogram of age
plt.figure(figsize=(8, 6))
sns.histplot(data['age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Boxplot of age by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='age', data=data)
plt.title('Age Distribution by Gender')
plt.show()

# 4. Scatterplot of age vs. year
plt.figure(figsize=(8, 6))
sns.scatterplot(x='year', y='age', data=data) # 'year' column is now available
plt.title('Age vs. Year')
plt.show()

# 5. Barplot of the number of deaths by race
plt.figure(figsize=(10, 6))
sns.countplot(x='race', data=data, order=data['race'].value_counts().index)
plt.title('Number of Deaths by Race')
plt.xticks(rotation=45)
plt.show()

# 6. Heatmap of correlation between numerical features
numerical_features = data.select_dtypes(include=np.number)
correlation_matrix = numerical_features.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# 7. Pairplot for selected numerical features
plt.figure(figsize=(10,8))
sns.pairplot(data[['age', 'year']], hue='year', diag_kind='kde')
plt.suptitle('Pairplot of Age and Year', y=1.02)
plt.show()
```

**Association for age and gender**

```
# prompt: Using dataframe data  association for  age and gender

import pandas as pd
import researchpy as rp

# Try 'latin-1' or 'ISO-8859-1' or 'cp1252' encoding
data = pd.read_csv('/content/PoliceKillingsUS.csv', encoding='latin-1')

# Convert age column to numeric, coercing errors to NaN
data['age'] = pd.to_numeric(data['age'], errors='coerce')

# Calculate the association between age and gender using Cramer's V
crosstab, res = rp.crosstab(data['age'], data['gender'], test='chi-square')
cramers_v = res.loc[2, 'results']

print(f"Cramer's V for the association between age and gender: {cramers_v}")

data['gender_numeric'] = data['gender'].astype('category').cat.codes

# Calculate the correlation
correlation = data['age'].corr(data['gender_numeric'])

print(f"Correlation between age and gender: {correlation}")

# Calculate the mean age for each gender
mean_age_by_gender = data.groupby('gender')['age'].mean()

print("\nMean age by gender:")
mean_age_by_gender
```
