# **Q.1. For 'House of Cards',what is the Show Id and Who is the Director of this show ?**

**isin()**

```python
data.head(2)
```

```python
# isin() is a method to know the element is present or not.

data[ data['Title'].isin(['House of Cards'])]
```

**str.contains()**

```python
#These methods are similar to check the required string present or not
data[data['Title'].str.contains('House of Cards')]
```

```python
ShowId = data[data['Title'].str.contains('House of Cards')]
ShowId['Show_Id']
```
