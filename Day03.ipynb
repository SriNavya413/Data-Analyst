```python
import pandas as pd

# Assuming your dataframe is named 'data'

# Calculate mean, median, and variance
numerical_data = data.select_dtypes(include=['number'])  # Select numerical columns
descriptive_stats = numerical_data.agg(['mean', 'median', 'var'])

# Calculate mode
mode = numerical_data.mode().iloc[0]  # Get the first mode for each column

# Calculate range
data_range = numerical_data.max() - numerical_data.min()


# Display the results
print("Descriptive Statistics:\n", descriptive_stats)
print("\nMode:\n", mode)
print("\nRange:\n", data_range)
```

```python
import pandas as pd
import numpy as np

# Assuming your dataframe is named 'data'

# Calculate mean, median, variance, standard deviation, and skewness
numerical_data = data.select_dtypes(include=['number'])
descriptive_stats = numerical_data.agg(['mean', 'median', 'var', 'std', 'skew'])

# Calculate mode
mode = numerical_data.mode().iloc[0]

# Calculate range
data_range = numerical_data.max() - numerical_data.min()

# Detect outliers using IQR method
Q1 = numerical_data.quantile(0.25)
Q3 = numerical_data.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numerical_data < (Q1 - 1.5 * IQR)) | (numerical_data > (Q3 + 1.5 * IQR))).any(axis=1)


# Display the results
print("Descriptive Statistics:\n", descriptive_stats)
print("\nMode:\n", mode)
print("\nRange:\n", data_range)
print("\nOutliers:\n", data[outliers])
```
