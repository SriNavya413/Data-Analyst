```
# Line graph
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```
# Scatter plot
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```
# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=20)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```
# Pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```
import seaborn as sns
```

```
import matplotlib.pyplot as plt
```

```
# Scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data)
plt.title('Scatter Plot')
plt.show()
```

```
# Line plot
sns.lineplot(x='sepal_length', y='sepal_width', hue='species', data=data)
plt.title('Line Plot')
plt.show()
```

```
# Histogram
sns.histplot(x='sepal_length', data=data, kde=True)
plt.title('Histogram')
plt.show()
```

```
# Box plot
sns.boxplot(x='species', y='sepal_length', data=data)
plt.title('Box Plot')
plt.show()
```

```
# Heatmap
corr_matrix = data.select_dtypes(include=np.number).corr() # Select only numerical columns for correlation
sns.heatmap(corr_matrix, annot=True)
plt.title('Heatmap')
plt.show()
```

```
# Pair plot
sns.pairplot(data, hue='species')
plt.title('Pair Plot')
plt.show()
```

```
# Joint plot
sns.jointplot(x='sepal_length', y='sepal_width', data=data, kind='kde')
plt.title('Joint Plot')
plt.show()
```

```
from scipy.stats import binom
```

```
# Set the parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success

# Generate x-values (number of successes)
x = np.arange(0, n + 1)

# Calculate the probability mass function (PMF)
pmf = binom.pmf(x, n, p)

# Create the bar plot
plt.bar(x, pmf)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n={}, p={})'.format(n, p))
plt.show()
```
