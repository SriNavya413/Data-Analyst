**df.loc[0]: This selects the row with the label 0 (the first row) from the DataFrame**

```python
import pandas as pd

# Sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 28],
        'City': ['New York', 'London', 'Paris']}
df = pd.DataFrame(data)

# Access the first row using df.loc[0]
first_row = df.loc[0]

# Print the first row
print(first_row)
```

**df.iloc[1]: This selects the row at index position 1 (the second row, because indexing starts at 0).**

```python
import pandas as pd

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

# Creating DataFrame
df = pd.DataFrame(data)

# Accessing the second row (index 1) using iloc
second_row = df.iloc[1]

# Print the second row
print(second_row)
```

**df.loc[0] accesses the row with label 0, which is the first row in this case.**
**df.iloc[1] accesses the row by position 1, which is the second row.**


```python
import pandas as pd

# Example DataFrame with custom index
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

# Creating DataFrame
df = pd.DataFrame(data)

# Displaying the original DataFrame
print("Original DataFrame:")
print(df)

# Accessing the first row using .loc (label-based indexing)
# Here, '0' is considered the label of the first row in the default integer index
loc_first_row = df.loc[0]
print("\nUsing df.loc[0] (label-based):")
print(loc_first_row)

# Accessing the second row using .iloc (position-based indexing)
# Here, '1' refers to the second row by its integer position
iloc_second_row = df.iloc[1]
print("\nUsing df.iloc[1] (position-based):")
print(iloc_second_row)
```

**axis=1: The axis=1 argument is used to indicate that you want to drop columns (not rows)**

```python
import pandas as pd

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago'],
    'Country': ['USA', 'USA', 'USA']
}

# Creating DataFrame
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Dropping a single column 'City' using axis=1
df_dropped_single = df.drop('City', axis=1)

# Display the DataFrame after dropping the 'City' column
print("\nDataFrame after dropping the 'City' column:")
print(df_dropped_single)

# Dropping multiple columns ['Age', 'Country'] using axis=1
df_dropped_multiple = df.drop(['Age', 'Country'], axis=1)

# Display the DataFrame after dropping multiple columns
print("\nDataFrame after dropping 'Age' and 'Country' columns:")
print(df_dropped_multiple)
```

**axis=0: The axis=0 argument tells pandas that you want to drop rows (not columns).**

```python
import pandas as pd

# Example DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

# Creating DataFrame
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Dropping a single row (at index 1) using axis=0
df_dropped_single = df.drop(1, axis=0)

# Display the DataFrame after dropping row at index 1
print("\nDataFrame after dropping the row at index 1:")
print(df_dropped_single)

# Dropping multiple rows (at indices 0 and 2) using axis=0
df_dropped_multiple = df.drop([0, 2], axis=0)

# Display the DataFrame after dropping rows at indices 0 and 2
print("\nDataFrame after dropping rows at indices 0 and 2:")
print(df_dropped_multiple)
```

**Use axis=1 when you want to remove columns from the DataFrame (e.g., removing a specific feature or unwanted column).**

**Use axis=0 when you want to remove rows from the DataFrame (e.g., removing a particular data entry or row).**

```python
# Dropping a column 'Age' (axis=1) and row at index 0 (axis=0)
df_dropped_column_row = df.drop('Age', axis=1).drop(0, axis=0)

print("DataFrame after dropping 'Age' column and row at index 0:")
print(df_dropped_column_row)
```

**axis=1 dropped the 'Age' column.**
**axis=0 dropped the row at index 0 **bold text** **
