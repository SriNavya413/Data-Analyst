```
import numpy as np
import scipy.stats as stats

# Sample data for A/B test
# A = Control group, B = Variant group
data = {
    "A": {"visitors": 1000, "conversions": 150},
    "B": {"visitors": 1200, "conversions": 180},
}

# Extract data
visitors_A = data["A"]["visitors"]
conversions_A = data["A"]["conversions"]
visitors_B = data["B"]["visitors"]
conversions_B = data["B"]["conversions"]

# Calculate conversion rates
rate_A = conversions_A / visitors_A
rate_B = conversions_B / visitors_B

# Print conversion rates
print(f"Conversion Rate A: {rate_A:.2%}")
print(f"Conversion Rate B: {rate_B:.2%}")

# Two-proportion z-test
# Null hypothesis: Conversion rates are equal (no difference)
# Alternative hypothesis: Conversion rates are different

# Calculate pooled conversion rate
pooled_rate = (conversions_A + conversions_B) / (visitors_A + visitors_B)
std_error = np.sqrt(pooled_rate * (1 - pooled_rate) * (1/visitors_A + 1/visitors_B))

# Calculate z-score
z_score = (rate_B - rate_A) / std_error

# Calculate p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Print results
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between A and B.")
else:
    print("Fail to reject the null hypothesis: No significant difference between A and B.")
```
