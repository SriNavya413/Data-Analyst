```python
# prompt: write a code for Generate a price_per_km column to assess cost efficiency.

# Assuming 'data5' is your DataFrame and it has 'price' and 'km' columns
if 'price' in data.columns and 'km' in data.columns:
    # Avoid division by zero by replacing zeros in 'km' with a small value (e.g., 1)
    # Ensure 'km' is numeric
    # data5['km'] = pd.to_numeric(data5['km'], errors='coerce').fillna(0).astype(int)

    data['price_per_km'] = data['price'] / data['km']
    print(data[['price', 'km', 'price_per_km']])
else:
    print("Error: 'price' or 'km' column not found in the DataFrame.")
```

# **Task-13:-**

```python
# prompt:  write code for Create categories for cars based on price (e.g., low, mid, high).
import matplotlib.pyplot as plt
import seaborn as sns


low_threshold = data['price'].quantile(0.33)
mid_threshold = data['price'].quantile(0.66)

# Function to categorize cars based on price
def categorize_price(price):
    if price <= low_threshold:
        return 'low'
    elif price <= mid_threshold:
        return 'mid'
    else:
        return 'high'

# Create the 'price_category' column
data['price_category'] = data['price'].apply(categorize_price)

# Display the DataFrame with the new category
print(data[['price', 'price_category']].head())

# Visualize the distribution of price categories (optional)
plt.figure(figsize=(8, 6))
sns.countplot(x='price_category', data=data)
plt.title('Distribution of Price Categories')
plt.xlabel('Price Category')
plt.ylabel('Number of Cars')
plt.show()
```

```python
# prompt: write code for Group the vehicles based on attributes like price, km, and age.

import pandas as pd

# Assuming 'data' is your DataFrame from the previous code
if 'price' in data.columns and 'km' in data.columns and 'car_age' in data.columns:
    # Group by price, km, and car_age and calculate the count of vehicles in each group
    grouped_vehicles = data.groupby(['price', 'km', 'car_age']).size().reset_index(name='vehicle_count')

    # Sort the groups by vehicle count in descending order to see the most common groups first
    grouped_vehicles = grouped_vehicles.sort_values('vehicle_count', ascending=False)

    print(grouped_vehicles)
else:
    print("Error: 'price', 'km', or 'car_age' column not found in the DataFrame.")
```
