# **TASK-2**


*** Standardize numerical fields (e.g., price, km) for consistency. ***


```python
# prompt: write a code for Standardize numerical fields (e.g., price, km) for consistency.

# Assuming 'data1' is your DataFrame and you want to standardize 'price' and 'km'
numerical_cols = ['price', 'km']

for col in numerical_cols:
    if col in data1.columns:  # Check if the column exists
        # Check if the column contains string values
        if data1[col].dtype == 'object':
            try:
                # Try to remove non-numeric characters and convert to numeric
                data1[col] = pd.to_numeric(data1[col].str.replace(r'[^0-9.]', '', regex=True))
            except ValueError:
                print(f"Warning: Column '{col}' contains non-numeric strings that cannot be automatically converted. Please review the data.")
                continue # Skip this column

        # Calculate mean and standard deviation
        mean = data1[col].mean()
        std = data1[col].std()

        # Handle cases where standard deviation is zero to avoid division by zero
        if std == 0:
            print(f"Warning: Standard deviation for column '{col}' is zero. Skipping standardization for this column.")
            continue  # Skip standardization for this column

        # Standardize the column
        data1[col] = (data1[col] - mean) / std
    else:
        print(f"Warning: Column '{col}' not found in DataFrame.")

# Display the DataFrame with standardized columns
data1
```

# **TASK-14**

***Create categories for cars based on price (e.g., low, mid, high).***

```python
# prompt: write code for Create categories for cars based on price (e.g., low, mid, high).

# Assuming 'data1' is your DataFrame and it has a 'price' column
if 'price' in data1.columns:
    # Define price thresholds for categories (adjust these as needed)
    low_threshold = data1['price'].quantile(0.33)  # 33rd percentile
    high_threshold = data1['price'].quantile(0.66) # 66th percentile

    # Create the 'price_category' column based on price thresholds
    data1['price_category'] = pd.cut(data1['price'], bins=[-float('inf'), low_threshold, high_threshold, float('inf')],
                                     labels=['low', 'mid', 'high'], right=False)

    print(data1[['price', 'price_category']].head())
else:
    print("Error: 'price' column not found in the DataFrame.")
```
